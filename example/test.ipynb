```python
import time

import cv2
import pytesseract

pytesseract.pytesseract.tesseract_cmd = "D:/install/ocr/tesseract.exe"

# 读取图像
img = cv2.imread("3.png")
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# 二值化
start = time.time()
_, binary = cv2.threshold(gray, 100, 255, cv2.THRESH_BINARY)

# 识别
config = "--psm 7 --oem 1 -c tessedit_char_whitelist=0123456789?=+-x÷"
text = pytesseract.image_to_string(binary, lang="eng", config=config)

print(text)
end = time.time()
print("耗时：", end - start, "s")
import matplotlib.pyplot as plt

img_raw = plt.imread("3.png")
plt.imshow(img_raw)
plt.figure()
plt.imshow(binary, cmap="gray")
```
